```python
import pandas as pd 
import numpy as np
```

```python
# Recuperation des fichiers des vehicules neufs commercialisés en france entre 2010 et 2018
df=pd.read_csv('data2010.csv')
df1=pd.read_csv('data2011.csv')
df2=pd.read_csv('data2012.csv')
df3=pd.read_csv('data2013.csv')
df4=pd.read_csv('data2014.csv')
df5=pd.read_csv('data2015.csv')
df6=pd.read_csv('data2016.csv')
df7=pd.read_csv('data2017.csv')
df8=pd.read_csv('data2018.csv')
df
```

```python
#Supprimer les colonnes inutiles des df
df=df.drop(['m (kg)','ep (KW)','Va', 'T','ID','Country','ec (cm3)','Erwltp (g/km)','Ct','Fm','IT','Status','VFN','Mp','Mh','Man','MMS'
            ,'Tan','Ve','Cr','Mt','Ewltp (g/km)','W (mm)','At1 (mm)','At2 (mm)','De','Vf'], axis='columns')
df1=df1.drop(['m (kg)','ep (KW)','Va', 'T','ID','Country','ec (cm3)','Erwltp (g/km)','Ct','Fm','IT','Status','VFN','Mp','Mh','Man','MMS'
              ,'Tan','Ve','Cr','Mt','Ewltp (g/km)','W (mm)','At1 (mm)','At2 (mm)','De','Vf'], axis='columns')
df2=df2.drop(['m (kg)','ep (KW)','Va', 'T','ID','Country','ec (cm3)','Erwltp (g/km)','Ct','Fm','IT','Status','VFN','Mp','Mh','Man','MMS'
              ,'Tan','Ve','Cr','Mt','Ewltp (g/km)','W (mm)','At1 (mm)','At2 (mm)','De','Vf'], axis='columns')
df3=df3.drop(['m (kg)','ep (KW)','Va', 'T','ID','Country','ec (cm3)','Erwltp (g/km)','Ct','Fm','IT','Status','VFN','Mp','Mh','Man','MMS'
              ,'Tan','Ve','Cr','Mt','Ewltp (g/km)','W (mm)','At1 (mm)','At2 (mm)','De','Vf'], axis='columns')
df4=df4.drop(['m (kg)','ep (KW)','Va', 'T','ID','Country','ec (cm3)','Erwltp (g/km)','Ct','Fm','IT','Status','VFN','Mp','Mh','Man','MMS'
              ,'Tan','Ve','Cr','Mt','Ewltp (g/km)','W (mm)','At1 (mm)','At2 (mm)','De','Vf'], axis='columns')
df5=df5.drop(['m (kg)','ep (KW)','Va', 'T','ID','Country','ec (cm3)','Erwltp (g/km)','Ct','Fm','IT','Status','VFN','Mp','Mh','Man','MMS'
              ,'Tan','Ve','Cr','Mt','Ewltp (g/km)','W (mm)','At1 (mm)','At2 (mm)','De','Vf'], axis='columns')
df6=df6.drop(['m (kg)','ep (KW)','Va', 'T','ID','Country','ec (cm3)','Erwltp (g/km)','Ct','Fm','IT','Status','VFN','Mp','Mh','Man','MMS'
              ,'Tan','Ve','Cr','Mt','Ewltp (g/km)','W (mm)','At1 (mm)','At2 (mm)','De','Vf'], axis='columns')
df7=df7.drop(['m (kg)','ep (KW)','Va', 'T','ID','Country','ec (cm3)','Erwltp (g/km)','Ct','Fm','IT','Status','VFN','Mp','Mh','Man','MMS'
              ,'Tan','Ve','Cr','Mt','Ewltp (g/km)','W (mm)','At1 (mm)','At2 (mm)','De','Vf'], axis='columns')
df8=df8.drop(['m (kg)','ep (KW)','Va', 'T','ID','Country','ec (cm3)','Erwltp (g/km)','Ct','Fm','IT','Status','VFN','Mp','Mh','Man','MMS'
              ,'Tan','Ve','Cr','Mt','Ewltp (g/km)','W (mm)','At1 (mm)','At2 (mm)','De','Vf'], axis='columns')
```

```python
# Concatener les df dans un seul df 
dfs=pd.concat([df,df1,df2,df3,df4,df5,df6,df7,df8])
dfs
```

```python
# Renommer les colonnes du df final
dfs=dfs.rename(columns = {'Mk':'marque','Cn':'modele','r': 'inscriptions'
                          ,'Enedc (g/km)':'co2','Ft':'carburant'
                          ,'z (Wh/km)':'consommation_energie_electrique_(wh/km)',
                          'Ernedc (g/km)':'reduction_de_emissions_grace_au_technologies_innovantes_(g/km)'
                          , 'year':'annee' })
dfs
```

```python
# Reindexer les colonnes du df
dfs=dfs.reindex(columns=['marque','modele','inscriptions','co2','carburant'
                         ,'consommation_energie_electrique_(wh/km)','reduction_de_emissions_grace_au_technologies_innovantes_(g/km)','annee'])
dfs
```

```python
# Remplacer les valeurs NaN par des 0
dfs['consommation_energie_electrique_(wh/km)'] = np.where(pd.isnull(dfs['consommation_energie_electrique_(wh/km)']),0,dfs['consommation_energie_electrique_(wh/km)'] )
dfs['reduction_de_emissions_grace_au_technologies_innovantes_(g/km)'] = np.where(pd.isnull(dfs['reduction_de_emissions_grace_au_technologies_innovantes_(g/km)']),0,dfs['reduction_de_emissions_grace_au_technologies_innovantes_(g/km)'] )
dfs
```

```python
dfs.to_csv('data2010to2018.csv')
```
