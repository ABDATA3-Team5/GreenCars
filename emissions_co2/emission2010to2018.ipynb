```python
import pandas as pd 
import numpy as np
```

```python
# Recuperation des fichiers des vehicules neufs commercialisés en france entre 2010 et 2018 
# et des fichiers des vehicules les plus commercilaisé en france
df=pd.read_csv('data2010to2018.csv')
df1=pd.read_csv('pdffinal.csv')
```

```python
df1.head()
```

```python
# Supprimer les colonnes inutiles
del df['Unnamed: 0']
del df1['Unnamed: 0']
```

```python
df1.annee.unique()
```

```python
#Selectionner les modéles des vehicules les plus commercialisés en france selon l'année 
filtered_df01= df1.query('annee == 2010')
values01=filtered_df01.modele
filtered_df02= df1.query('annee == 2011')
values02=filtered_df02.modele
filtered_df03= df1.query('annee == 2012')
values03=filtered_df03.modele
filtered_df04= df1.query('annee == 2013')
values04=filtered_df04.modele
filtered_df05= df1.query('annee == 2014')
values05=filtered_df05.modele
filtered_df06= df1.query('annee == 2015')
values06=filtered_df06.modele
filtered_df07= df1.query('annee == 2016')
values07=filtered_df07.modele
filtered_df08= df1.query('annee == 2017')
values08=filtered_df08.modele
filtered_df09= df1.query('annee == 2018')
values09=filtered_df09.modele
```

```python
#  Creation des DF a partir de la selection des modéles des vehicules commercialisés en France 
#selon les modéles les plus vendus par rapport à chaque année
filtered_df11 = df[df.modele.isin(values01)&df.annee.eq(2010)]
filtered_df12 = df[df.modele.eq('206')&df.annee.eq(2010)]
filtered_df13 = df[df.modele.eq('207')&df.annee.eq(2010)]
filtered_df14 = df[df.modele.eq('C4')&df.annee.eq(2010)]
filtered_df15 = df[df.modele.eq('XSARA PICASSO')&df.annee.eq(2010)]
filtered_df01=pd.concat([filtered_df11,filtered_df12,filtered_df13,filtered_df14,filtered_df15])

filtered_df21 = df[df.modele.isin(values02)&df.annee.eq(2011)]
filtered_df22 = df[df.modele.eq('206')&df.annee.eq(2011)]
filtered_df23 = df[df.modele.eq('207')&df.annee.eq(2011)]
filtered_df02=pd.concat([filtered_df21,filtered_df22,filtered_df23])

filtered_df31 = df[df.modele.isin(values03)&df.annee.eq(2012)]
filtered_df32 = df[df.modele.eq('206')&df.annee.eq(2012)]
filtered_df33 = df[df.modele.eq('207')&df.annee.eq(2012)]
filtered_df34 = df[df.modele.eq('208')&df.annee.eq(2012)]
filtered_df03=pd.concat([filtered_df31,filtered_df32,filtered_df33,filtered_df34])

filtered_df41 = df[df.modele.isin(values04)&df.annee.eq(2013)]
filtered_df42 = df[df.modele.eq('206')&df.annee.eq(2013)]
filtered_df43 = df[df.modele.eq('207')&df.annee.eq(2013)]
filtered_df44 = df[df.modele.eq('208')&df.annee.eq(2013)]
modeles=[r'114(?!$)',r'116(?!$)',r'118(?!$)',r'120(?!$)',r'123(?!$)',r'125(?!$)',r'135(?!$)',
         r'518(?!$)',r'520(?!$)',r'525(?!$)',r'528(?!$)',r'530(?!$)',r'535(?!$)',r'550(?!$)',r'M5(?!$)']
values4=filtered_df04.marque
filtered_df4 = df[df.marque.isin(values4)&df.annee.eq(2013)]
filtered_df45=filtered_df4[filtered_df4['modele'].str.contains('|'.join(modeles))]
filtered_df04=pd.concat([filtered_df41,filtered_df42,filtered_df43,filtered_df44,filtered_df45])

filtered_df51 = df[df.modele.isin(values05)&df.annee.eq(2014)]
filtered_df52 = df[df.modele.eq('206')&df.annee.eq(2014)]
filtered_df53 = df[df.modele.eq('207')&df.annee.eq(2014)]
filtered_df54 = df[df.modele.eq('208')&df.annee.eq(2014)]
filtered_df05=pd.concat([filtered_df51,filtered_df52,filtered_df53,filtered_df54])

filtered_df06 = df[df.modele.isin(values06)&df.annee.eq(2015)]

filtered_df07 = df[df.modele.isin(values07)&df.annee.eq(2016)]

filtered_df08 = df[df.modele.isin(values08)&df.annee.eq(2017)]

filtered_df09 = df[df.modele.isin(values09)&df.annee.eq(2018)]
```

```python
# Concatenation des DF
filtered_df=pd.concat([filtered_df01,filtered_df02,filtered_df03,filtered_df04,filtered_df05,filtered_df06,filtered_df07,filtered_df08,filtered_df09])
```

```python
# Selectionner les modeles de la marque BMW  sur le df 
bmw = filtered_df[(filtered_df.marque == 'BMW')]
bmw['modele'].unique()
```

```python
# Regrouper les modeles de serie 1 et serie 5 dans des listes
one_series = ['120D XDRIVE', '116D', '114I', '118I',
       '125I', '118D', '125D', '114D', '116I', '120D', 'M135I',
       '118D XDRIVE', '116ED', '135I', '123D', '120I', 'M135I XDRIVE']

five_series = ['520D', '530D',
       '535D', '530D XDRIVE', '535I', '525D XDRIVE', '535D XDRIVE',
       '535I XDRIVE', '550I XDRIVE', 'M 550D XDRIVE', '525D', '528I',
       '528I XDRIVE', '518D', '520D XDRIVE','530I']
```

```python
# Renommer tout les modeles listé dans les listes en serie1 et serie 5 dans le df
filtered_df['modele'] = filtered_df['modele'].apply(lambda x:'SERIE 1' if x in one_series
                                                   else 'SERIE 5' if x in five_series
                                                    else 'X5' if x == 'X5 M50D'
                                                    else 'X6' if x == 'X6 M50D' else x)
filtered_df["carburant"] = filtered_df["carburant"].str.lower()
filtered_df.head()
```

```python
# Faire un groupby pour obtenir le minimum d'emission co2 pour chaque modele de vehicule en fonction du carburant et de l'année
final_df1 = filtered_df.groupby(["marque", "modele", "carburant", "annee"]).min()
final_df1=pd.DataFrame(final_df1)
final_df1=final_df1.reset_index()
final_df1=final_df1.rename(columns={'co2':'min_co2'})
final_df1 = final_df1[["marque", "modele", "carburant", "min_co2", "annee"]]
final_df1.head()
```

```python
# Faire un groupby pour obtenir la moyenne d'emission co2 pour chaque modele de vehicule en fonction du carburant et de l'année
final_df = filtered_df.groupby(["marque", "modele", "carburant", "annee"]).mean()
final_df=pd.DataFrame(final_df)
final_df=final_df.reset_index()
final_df=final_df.rename(columns={'co2':'moyenne_co2'})
final_df.head()
```

```python
# Faire un groupby pour obtenir la maximum d'emission co2 pour chaque modele de vehicule en fonction du carburant et de l'année
final_df2 = filtered_df.groupby(["marque", "modele", "carburant", "annee"]).max()
final_df2=pd.DataFrame(final_df2)
final_df2=final_df2.reset_index()
final_df2=final_df2.rename(columns={'co2':'max_co2'})
final_df2 = final_df2[["marque", "modele", "carburant", "max_co2", "annee"]]
final_df2.head()
```

```python
# Merge les 3 df qui contiennent le min, moyenne et max entre eux
merged_df_co2_1 = final_df.merge(final_df1, how = "left", on = ["marque", "modele", "annee","carburant"])
merged_df_co2= merged_df_co2_1.merge(final_df2, how = "left", on = ["marque", "modele", "annee","carburant"])
merged_df_co2=merged_df_co2.fillna(0)
merged_df_co2
```

```python
# Elliminer un e de la colonne rang du df1
df1.rang = df1.rang.str.strip('e')
df1["rang"] = df1["rang"].astype(float)
df1["rang"] = df1["rang"].astype(int)
df1.head()
```

```python
# Merge le merged_df_co2 avec df1 et reindexer les colonnes du dataframe final 
merged_df = merged_df_co2.merge(df1, how = "left", on = ["marque", "modele", "annee"])
merged_df=merged_df.reindex(columns=['marque','modele','carburant','annee','inscriptions','min_co2','moyenne_co2','max_co2','consommation_energie_electrique_(wh/km)','reduction_de_emissions_grace_au_technologies_innovantes_(g/km)','rang','volume','%_marche'])
merged_df
```

```python
# Transformation du type de la colonne marche en float
merged_df['%_marche'],merged_df['%0']=merged_df['%_marche'].str.split(',',2).str
merged_df['.'] = '.'
merged_df['%_marche']=merged_df['%_marche']+merged_df['.']+merged_df['%0']
del merged_df['%0']
del merged_df['.']
merged_df['%_marche'] = merged_df['%_marche'].apply(pd.to_numeric, errors='coerce')
merged_df
```

```python
# calculer l'ecrat type de la moyenne du Co2
ecar_moy=np.std(merged_df['moyenne_co2'])
ecar_moy
```
